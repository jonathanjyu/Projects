```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(3,2, figsize=(15, 15))
ax[0][0].scatter(df['sepal length (cm)'],df['sepal width (cm)'])
ax[0][1].scatter(df['sepal length (cm)'],df['petal length (cm)'])

ax[1][0].scatter(df['sepal length (cm)'],df['petal width (cm)'])
ax[1][1].scatter(df['sepal width (cm)'],df['petal length (cm)'])

ax[2][0].scatter(df['sepal width (cm)'],df['petal width (cm)'])
ax[2][1].scatter(df['petal length (cm)'],df['petal width (cm)'])
```

```python
# kind:非對角線圖形, diag_kind:對角線圖形, hue:依照class分群df['target']
sns.pairplot(df,kind="reg", hue='target', diag_kind='kde')
plt.show()
```

```python
# 設置 Seaborn 的樣式
sns.set(style="whitegrid")

# 繪製 Sepal 長度和寬度的散點圖
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', palette='viridis', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()

# 繪製 Petal 長度和寬度的散點圖
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', palette='viridis', data=df)
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
# 繪製箱型圖
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='target', y='sepal length (cm)', data=df)
plt.title('Sepal Length by Target')

plt.subplot(2, 2, 2)
sns.boxplot(x='target', y='sepal width (cm)', data=df)
plt.title('Sepal Width by Target')

plt.subplot(2, 2, 3)
sns.boxplot(x='target', y='petal length (cm)', data=df)
plt.title('Petal Length by Target')

plt.subplot(2, 2, 4)
sns.boxplot(x='target', y='petal width (cm)', data=df)
plt.title('Petal Width by Target')

plt.tight_layout()
plt.show()
```

```python
# 繪製小提琴圖
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.violinplot(x='target', y='sepal length (cm)', data=df)
plt.title('Sepal Length by Target')

plt.subplot(2, 2, 2)
sns.violinplot(x='target', y='sepal width (cm)', data=df)
plt.title('Sepal Width by Target')

plt.subplot(2, 2, 3)
sns.violinplot(x='target', y='petal length (cm)', data=df)
plt.title('Petal Length by Target')

plt.subplot(2, 2, 4)
sns.violinplot(x='target', y='petal width (cm)', data=df)
plt.title('Petal Width by Target')

plt.tight_layout()
plt.show()
```

```python
# 計算相關係數矩陣
correlation_matrix = df.iloc[:, :-1].corr()

# 顯示相關係數矩陣
print(correlation_matrix)

# 設置繪圖風格
sns.set(style='white')

# 繪製相關係數矩陣的熱力圖
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```
