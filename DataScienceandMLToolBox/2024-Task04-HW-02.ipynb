```python
import numpy as np
import pandas as pd
raw_data = {'first_name': ['Jason', np.nan, 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', np.nan, 'Ali', 'Milner', 'Cooze'], 
        'age': [42, np.nan, 36, 24, 73], 
        'sex': ['m', np.nan, 'f', 'm', 'f'], 
        'preTestScore': [4, np.nan, np.nan, 2, 3],
        'postTestScore': [25, np.nan, np.nan, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'sex', 'preTestScore', 'postTestScore'])
df
```

```python
##### 1.Drop missing observations
df1 = df.dropna()
df1
```

```python
##### 2.Drop rows where all cells in that row is NA
df2 = df.dropna(axis = 0, how='all')
df2
```

```python
##### 3.Create a new column full of missing values
df3 = df.copy()
df3['new_column'] = np.nan
df3
```

```python
##### 4.Fill in missing data with zeros
df4 = df.fillna(0)
df4
```

```python
##### 5.Fill in missing in preTestScore with the mean value of preTestScore
df5 = df.copy()
df5['preTestScore'].fillna(df5['preTestScore'].mean(), inplace=True)
df5
```

```python
##### 6.Fill in missing in postTestScore with each sex’s mean value of postTestScore
df6 = df.copy()
fillnam = df6[df6['sex']=='m']['postTestScore'].mean()
fillnaf = df6[df6['sex']=='f']['postTestScore'].mean()

df6.loc[(df6['sex']=='f')&(df6['postTestScore'].isnull()), 'postTestScore'] = fillnaf
df6.loc[(df6['sex']=='m')&(df6['postTestScore'].isnull()), 'postTestScore'] = fillnam
df6
```

```python
##### 7.Select some rows but ignore the missing data points
df7 = df.copy()
df7.dropna()
```
