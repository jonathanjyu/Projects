```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [
  ['Milk', 'Onion', 'Nutmeg', 'Kidney' 'Beans', 'Eggs', 'Yogurt'],
  ['Dill', 'Onion', 'Nutmeg', 'Kidney' 'Beans', 'Eggs', 'Yogurt'],
  ['Milk', 'Apple', 'Kidney' 'Beans', 'Eggs'],
  ['Milk', 'Unicorn', 'Corn', 'Kidney' 'Beans', 'Yogurt'],
  ['Corn', 'Onion', 'Onion', 'Kidney' 'Beans', 'Ice cream', 'Eggs'],
]
dataset = pd.DataFrame(dataset)
dataset
```

```python
set(dataset)
```

```python
item_set = set()
for i in range(len(dataset)):
    #print(set(dataset[i]))
    #if dataset[i] 
    item_set.update(set(dataset[i]))
print(item_set)
print(len(item_set))
```

```python
item_set = list(item_set)
```

```python
item_set = [i for i in item_set if i is not None]
item_set
```

```python
tranform_set = []
for i in range(len(item_set)):
    #print(item_set[i])
    tmp = []
    for j in range(len(dataset)):
        #print(list(dataset.iloc[j]))
        if item_set[i] in list(dataset.iloc[j]):
            tmp.append(1)
        else:
            tmp.append(0)
    tranform_set.append(tmp)
print(tranform_set)
```

```python
data = pd.DataFrame(tranform_set).T
data
```

```python
data.columns=item_set
data
```

```python
frequent_itemsets = apriori(data, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)

# 按照信賴度排序並輸出結果
rules.sort_values('confidence', ascending=False, inplace=True)
rules[['antecedents', 'consequents', 'support', 'confidence']]
```

# 整理後

```python
import pandas as pd
import numpy as np

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

dataset = [
  ['Milk', 'Onion', 'Nutmeg', 'Kidney' 'Beans', 'Eggs', 'Yogurt'],
  ['Dill', 'Onion', 'Nutmeg', 'Kidney' 'Beans', 'Eggs', 'Yogurt'],
  ['Milk', 'Apple', 'Kidney' 'Beans', 'Eggs'],
  ['Milk', 'Unicorn', 'Corn', 'Kidney' 'Beans', 'Yogurt'],
  ['Corn', 'Onion', 'Onion', 'Kidney' 'Beans', 'Ice cream', 'Eggs'],
]

print('dataset:\n',dataset)

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

print('\nafter transform:\n',df)

frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
print("\nFrequent Patterns（Apriori）:\n",frequent_itemsets)

rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
rules.sort_values('confidence', ascending=False, inplace=True)
print("\nAssociation Rules:\n",rules[['antecedents', 'consequents', 'support', 'confidence']])
```
