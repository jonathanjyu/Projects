```python
import numpy as np
import pandas as pd
url = 'https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv'
df = pd.read_csv(url)
df
```

```python
##### Count the number of missing values in each column and find the maximum number of missing values.
print('The number of missing values in each column:\n',df.isnull().sum())
print('\nThe maximum number of missing values:',max(df.isnull().sum()))
```

```python
##### How to replace missing values of `multiple numeric` columns with the mean?
df2 = df.copy()
df2.fillna(df2.mean(), inplace=True)
print('The number of missing values in each column:\n',df2.isnull().sum())
print('\nThe maximum number of missing values:',max(df2.isnull().sum()))
```

```python
##### Calculate the average price of different manufacturers.
print(df.groupby('Manufacturer')['Price'].mean())
```

```python
##### How to replace missing values of `price` columns with the mean depending on its manufacturers?
df3 = df.copy()
df3['Price'] = df3.groupby('Manufacturer')['Price'].transform(lambda x: x.fillna(x.mean()))
```

```python
print('The number of missing values in each column:\n',df3.isnull().sum())
print('\nThe maximum number of missing values:',max(df3.isnull().sum()))
```
